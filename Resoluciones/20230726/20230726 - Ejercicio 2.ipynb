Para la comprobación mediante software:

$$ S_{11} = \frac{Z}{2 R_{O1} Z} $$

$$ Z = \frac{2 R_{O1} S_{11}}{ 1 - S_{11} } $$ 

```python
from sympy import *
from pytc2.sistemas_lineales import pretty_print_SOS
import numpy as np


s = symbols('s')


R_O1 = 1
S_11 = (20*s**2 + 12*s +2)/(30*s**2 + 14*s + 4)

Z = (2 * R_O1 * S_11) / (1 - S_11)
Z.simplify()
```

```python
lim_Z_0 = limit(Z, s, 0)
print("El límite de Z cuando s tiende a cero es: {}".format(lim_Z_0))
print("Entonces, R1 = {}".format(lim_Z_0))
R_1 = lim_Z_0
```

```python
lim_Z_inf = limit(Z, s, oo)
print("El límite de Z cuando s tiende a infinito es: {}".format(lim_Z_inf))
print("Entonces, R1 + R2 = {}".format(lim_Z_inf))
print("R2 = {} - R1".format(lim_Z_inf))
R_2 = lim_Z_inf - R_1
print("R2 = {}".format(R_2))
```

Partiendo de $L = 10$

$ Z_2 = Z - R_1 $

```python
import sympy

Z_2 = Z - R_1


Z_2.simplify()
```

Definimos $Y_2$ como la inverza de $Z_2$

```python
Y_2 = 1 / Z_2

print("Y_2 = ")
Y_2.simplify()
```

Definimos $Y_L$ como la admitancia de la bobina

```python
print("Y_L = ")
Y_L = 1 / (10 * s)
Y_L.simplify()
```

Definimos $Y_4$ como $Y_2$ - $Y_L$

```python
Y_4 = Y_2 - Y_L
Y_4.simplify()
```

Obteniendo por resultado la suma serie de $R_2$ y $C$

```python
Z_4 = 1 / Y_4
Z_4.simplify()
```

Restamos $R_2$ a $Z_4$ para obtener $C$, entonces $C$ será: 

```python
C = Z_4 - R_2
C.simplify()
```

Conocemos:
