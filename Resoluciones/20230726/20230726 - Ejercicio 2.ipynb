Para la comprobacin mediante software:

$$ S_{11} = \frac{Z}{2 R_{O1} Z} $$

$$ Z = \frac{2 R_{O1} S_{11}}{ 1 - S_{11} } $$ 

```python
from sympy import *
from pytc2.sistemas_lineales import pretty_print_SOS
import numpy as np


s = symbols('s')


R_O1 = 1
S_11 = (20*s**2 + 12*s +2)/(30*s**2 + 14*s + 4)

Z = (2 * R_O1 * S_11) / (1 - S_11)
Z.simplify()
```

```python
lim_Z_0 = limit(Z, s, 0)
print("El lmite de Z cuando s tiende a cero es: {}".format(lim_Z_0))
print("Entonces, R1 = {}".format(lim_Z_0))
R_1 = lim_Z_0
```

```python
lim_Z_inf = limit(Z, s, oo)
print("El lmite de Z cuando s tiende a infinito es: {}".format(lim_Z_inf))
print("Entonces, R1 + R2 = {}".format(lim_Z_inf))
print("R2 = {} - R1".format(lim_Z_inf))
R_2 = lim_Z_inf - R_1
print("R2 = {}".format(R_2))
```
